```python
def func(x):
    return x ** 2 + 10 * np.sin(x)

x = np.linspace(-10, 10, 100)
y = func(x)

plt.plot(x, y, label = 'y = x^2 + 10sin(x)')

x_min = sp.optimize.fminbound(func, x1 = 2, x2 = 7.5)
y_min = func(x_min)

plt.plot(x_min, y_min, 'ro', label = 'Minimum @ ({0:.3f}, {1:.3f})'.format(x_min, y_min))
plt.axvline(2,   color = 'black')
plt.axvline(7.5, color = 'black')

plt.legend(loc = 'upper right')
plt.show()
```

```python
def func(x):
    return x ** 2 + 10 * np.sin(x)

x = np.arange(-10, 10, 0.1)
y = func(x)

plt.plot(x, y, label = 'y = x^2 + 10sin(x)')

a = sp.optimize.fmin_bfgs(func, 
                          x0 = 50,
                          epsilon = 0.1,
                          disp = 3,
                          retall = 1,
                          maxiter = 10)

a
```

```python
def func(x):
    return x ** 2 + 10 * np.sin(x)

x = np.arange(-10, 10, 0.1)
y = func(x)

plt.plot(x, y, label = 'y = x^2 + 10sin(x)')

a = sp.optimize.basinhopping(func, 
                              x0 = 50,
                              niter = 10,
                              stepsize = 0.1,
                              disp = 0)

x_min = a.x[0]
y_min = func(x_min)

plt.plot(x_min, y_min, 'ro', label = 'Minimum @ ({0:.3f}, {1:.3f})'.format(x_min, y_min))

plt.legend(loc = 'upper right')
plt.show()
```

### get the best fit line

```python
def func(x, a, b, c):
    return a * np.exp(-b * x) + c

xdata = np.linspace(0, 5, 50)
y = func(xdata, a = 2.5, b = 1.3, c = 0.5)

ydata = y + 0.2 * np.random.normal(size = len(xdata))
np.random.seed(1729)

plt.plot(xdata, ydata, label = 'data')

popt, pcov = sp.optimize.curve_fit(func, xdata, ydata)

plt.plot(xdata, func(xdata, *popt), label = 'fit line')

plt.legend()
plt.show()
```
