## Data Normalization (Scaling)

![](https://miro.medium.com/v2/resize:fit:720/format:webp/1*tNRMMFhNcYdOP7KebtTfSA.png)

### Why ?
1. Important for distance-based algorithms :
    - K-means - Clustering
    - K nearest neighbours (KNN) - Classification
    - Principal Component Analysis (PCA) - Dimensionality reduction
    > Without it the features with large ranges will **dominate**
2. Important for gradient-based algorithms :
    - Linear/Non-linear Regression
    - Logistic Regression
    - Neural Networks
    > Without it training will become **slow and may oscillate**

However some algorithms work fine without feature scalling **because** they *work on each feature independently*:
- Decision Trees
- Naive Bayes Classifier

### When ?
- After Data Spliting the dataset to (training - validation - testing) on taining input **ONLY** - not entire dataset

### How ?

#### Using Standardization

$$\large{X_{new} = \frac{X_{old} - \mu}{σ}}$$
where :
- $\mu$ : mean
- $σ$ : Standard Deviation

```python
data = pd.read_csv('/content/drive/MyDrive/Data-examples/example8.csv')
data_input  = data.drop('y', axis = 1)   # columns = ['y']
data_output = data['y']
x_train, x, y_train, y = train_test_split(data_input, data_output, test_size = 0.3, random_state = 42)
x_test, x_val, y_test, y_val = train_test_split(x, y, test_size = 0.5, random_state = 42)
data_input.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)

x_val_scaled = scaler.fit_transform(x_val)
x_test_scaled = scaler.fit_transform(x_test)

'''
or

x_train_scaled = scaler.fit_transform(x_train)

'''
x_train_scaled[:10]
```

```python
x_train_scaled = pd.DataFrame(x_train_scaled, columns = x_train.columns)
x_train_scaled.head()
```
