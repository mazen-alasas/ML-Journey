```python
# function to create a dataframe

def create_df(col, row):
    d = {c: [str(c) + str(i) for i in row] for c in col}
    return pd.DataFrame(d, index = row)

df = create_df(['a', 'b', 'c'], [1, 2, 3])
# or
df = create_df('abc', range(1, 4))

df
```

```python
# merge dataframes
# there must be a common column in both dataframes to merge
# does not work if there is no common column

df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'employee': ['Lisa', 'Bob', 'Jake', 'Sue'],
                    'hire_date': [2004, 2008, 2012, 2014]})

df3 = pd.merge(df1, df2)

df3
```

```python
df1 = pd.DataFrame({'a': [1, 2, 3, 4],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'hire_date': [2004, 2008, 2012, 2014],
                    'a': [1, 2, 3, 4]})

df3 = pd.merge(df1, df2)

df3
```

```python
df1 = pd.DataFrame({'employee': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'group': ['Accounting', 'Engineering', 'Engineering', 'HR']})
df2 = pd.DataFrame({'name': ['Bob', 'Jake', 'Lisa', 'Sue'],
                    'salary': [70000, 80000, 120000, 90000]})

df = pd.merge(df1, df2, left_on = "employee", right_on = "name").drop('name', axis = 1)

df
```

```python
d = pd.DataFrame({'key' : ['A', 'B', 'C', 'A', 'B', 'C'],
                  'data': range(6)}, columns=['key', 'data'])

df = d.groupby('key').sum()
df
```

```python
d  = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': np.random.randint(0, 10, 6)})

d
```

```python
df = d.groupby('key').aggregate({'data1': 'min', 'data2': 'max'})

df
```

```python
def filter_func(x):
    return x['data2'].std() > 4

df = d.groupby('key').filter(filter_func)

df
```

```python
df = d.groupby('key').transform(lambda x: x**2)

df
```
