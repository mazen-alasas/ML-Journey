```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size = 1000)

sns.kdeplot(x = data[:, 0], y = data[:, 1])
# plt.show()
```

```python
sns.kdeplot(x = 'x', y = 'y', data = data, fill = True) 
```

```python
sns.kdeplot(x = data[:, 0], y = data[:, 1], shade = True)
```

```python
sns.kdeplot(x = data[:, 0], y = data[:, 1], shade = True, cmap = 'Reds')
```

```python
data = pd.DataFrame(data, columns = ['x', 'y'])

sns.kdeplot(data)
```

```python
for col in data.columns:
    sns.kdeplot(data[col], shade = True)
```

```python
for col in data.columns:
    sns.kdeplot(data[col], fill = True)
```

```python
with sns.axes_style("white"):
    sns.jointplot(x = 'x', y = 'y', data = data, kind = 'kde', fill = True)
```

```python
with sns.axes_style("white"):
    sns.jointplot(x = 'x', y = 'y', data = data, kind = 'hex')
```
