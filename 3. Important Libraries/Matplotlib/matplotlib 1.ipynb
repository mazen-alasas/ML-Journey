```python
import csv
from collections import Counter

with open('data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    languages_ctr = Counter()

    for row in csv_reader:
        languages_ctr.update(row['LanguagesWorkedWith'].split(';'))

languages_count = languages_ctr.most_common(10)

languages = []
popularity = []

for item in languages_count:
    languages.append(item[0])
    popularity.append(item[1])
```

```python
plt.bar(languages, popularity)

plt.title("Most popular languages")
plt.xlabel("Languages")
plt.ylabel("Popularity")
```

```python
plt.barh(languages, popularity)

plt.title("Most popular languages")
plt.xlabel("Popularity")
plt.ylabel("Languages")
```

```python
languages.reverse()
popularity.reverse()

plt.barh(languages, popularity)

plt.title("Most popular languages")
plt.xlabel("Popularity")
plt.ylabel("Languages")
```

```python
import pandas as pd

df = pd.read_csv('data.csv')
ids = df['Responder_id']
lang_rows = df['LanguagesWorkedWith']

language_counter = Counter()

for row in lang_rows:
    language_counter.update(row.split(';'))

languages = []
popularity = []

for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

languages.reverse()
popularity.reverse()

plt.barh(languages, popularity)
plt.title("Most popular languages")
plt.xlabel("Popularity")
plt.ylabel("Languages")
```

```python
# Colors:
Blue = '#008fd5'
Red = '#fc4f30'
Yellow = '#e5ae37'
Green = '#6d904f'

slices = [60,       40,      30,       20]
labels = ['Sixty', 'Forty', 'Thirty', 'Twenty']
colors = [Blue,     Red,     Yellow,   Green]
plt.pie(slices, labels = labels, 
        colors = colors, 
        wedgeprops = {'edgecolor': 'black'})
plt.show()
```

```python
plt.pie(popularity, labels = languages)
plt.show()
```

```python
explode = [0, 0, 0.1, 0, 0.1, 0, 0]

plt.pie(popularity, 
        labels = languages, 
        explode = explode,
        shadow = True)
plt.show()
```

```python
plt.pie(popularity, 
        labels = languages, 
        explode = explode,
        shadow = True,
        startangle = 90)
plt.show()
```

```python
plt.pie(popularity, 
        labels = languages, 
        explode = explode,
        autopct = '%1.1f%%',
        startangle = 90)
plt.show()
```
