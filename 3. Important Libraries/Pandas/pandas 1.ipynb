```python
d = pd.Series([45, 56, 67, 78, 89, 45, 56, 55, 45, 56, 67, 78, 89, 67, 78, 89])

d.describe()
```

```python
print(d.agg(['sum']), end = "\n\n")

data_min = d.min()
print(data_min, end = "\n\n")

print(d.agg(['sum', 'mean', 'std']))
```

```python
d1 = pd.Index([45, 56, 67, 78, 89, 45])
d2 = pd.Index([47, 78, 89, 67, 78, 89])

print(d1 & d2)                   # and between each element and the corresponding to him
print(d1 | d2)                   # or // /// /// / / // //// / / // / / // / / / / / / /
print(d1 ^ d2)                   # xor // / ////        / // / /////// // //// //////  /
```

```python
d1 = pd.Index([45, 56, 67, 78, 89, 45])
d2 = pd.Index([47, 78, 89, 67, 78, 89])

print(d1.intersection(d2))
print(d1.union(d2))
print(d1.difference(d2))
print(d1.symmetric_difference(d2))
```

```python
d1 = pd.Index([45, 56, 67, 78, 89, 45])
d2 = pd.Index([47, 78, 89, 67, 78, 89])

print(d1.isin(d2))               # check if all elements in d1 is in d2
print(d2.isin(d1))
```
